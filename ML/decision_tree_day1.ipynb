```python
import pandas as pd
fruits = pd.read_csv('../pd/DataSets/fruits.csv')
X = fruits[['sphericity','weight']]
y = fruits['label']
fruits.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x='sphericity', y='weight',data=X[y=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity', y='weight', data=X[y=='Orange'], c='green', label='Orange')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X,y) # Build the decision tree classifier from the training set (X,y)
plt.figure(figsize=[8,8])
plot_tree(model, feature_names=['sphericity','weight'], class_names=['Apple','Orange'])
plt.show()
```
