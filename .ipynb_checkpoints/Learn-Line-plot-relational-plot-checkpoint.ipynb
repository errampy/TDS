```python
months = ['jan', 'feb', 'march', 'april', 'may','june', 'july']
sales = [1000,1300,1500,1800,1200,2500,2700]
plt.plot(months, sales, marker='o',ms=10,ls=':', lw=2, c='red', mfc='blue',mec='yellow', mew=5)
# ms ==> marker size,
# ls ==> line style
# lw ==> line width
# mfc ==> marker face color
# mec ==> marker edge color , its use to do border of marker color
# mew ==> marker edge width
plt.show()
```
