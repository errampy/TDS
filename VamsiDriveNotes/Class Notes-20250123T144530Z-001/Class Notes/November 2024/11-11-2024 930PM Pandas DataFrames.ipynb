```python
def convert_species(s):
    return " ".join(s.split("-"))
```

```python
convert_species('Iris-virginica')
```

```python
iris = pd.read_csv('e:/data/iris.csv',
                   header=None,
                   names=['sepal_length','sepal_width','petal_length','petal_width','species'],
                   converters={'species':convert_species}
                  )
iris
```
