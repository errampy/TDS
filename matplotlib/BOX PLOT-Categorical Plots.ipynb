# Categorical Plots

its mainly Use to understands the relationship between 2 variable is called relational plots
Mainly used to ananly the some kinds of categorical data is called categorical plots
1. Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
# box plot is univariate analsyss i
cars = pd.read_csv('../pd/DataSets/usedcars.csv')
cars.head()
```

```python
plt.boxplot(x='price',data=cars)
plt.show()
```

```python
# change outlayer symple
plt.boxplot(x='price', data=cars, sym='d')
'''
sym = symbo
s - square
*
d = diamond
^  up tranles
'''
plt.show()
```

```python
# Don't show the out layer
plt.boxplot(x='price', data=cars,showfliers=False)
plt.show()
```

```python
#

# wiskers default value is 1.5
# How much value you wants to multy the IQR Value,
plt.boxplot(x='price', data=cars, whis=1.5)
plt.show()
```

```python
# how to highligh the mean value
plt.boxplot(x='price', data=cars, notch=True)
plt.show()
```

```python
# how to change vertical to horizontal,
plt.boxplot(x='price', data=cars, vert=False)
plt.show()
```

```python
# how to control width of box
plt.boxplot(x='mileage', data=cars, widths=0.4)
plt.show()
# Note :  we can't control height.
```

```python
plt.boxplot(x='price', data=cars[cars.model=='SE'])
plt.boxplot(x='price', data=cars[cars.model=='SES'])
plt.boxplot(x='price', data=cars[cars.model=='SEL'])
plt.show()
```

```python
# How to set the position of box , defeault are all 0
```

```python
plt.boxplot(x='price', data=cars[cars.model=='SE'], positions=[0], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SES'], positions=[1], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SEL'], positions=[2], widths=0.4)
plt.show()
```

```python
plt.boxplot(x='price', data=cars[cars.model=='SE'], positions=[0], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SES'], positions=[1], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SEL'], positions=[2], widths=0.4)
plt.xlabel('Cars model')
plt.ylabel('Cars price')
plt.title('Box plot for cars prices')
plt.xticks(ticks=[0,1,2],labels=['SE','SES','SEL'])
plt.show()
```

```python
# How to set and rorate the x-ticks
plt.boxplot(x='price', data=cars[cars.model=='SE'], positions=[0], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SES'], positions=[1], widths=0.4)
plt.boxplot(x='price', data=cars[cars.model=='SEL'], positions=[2], widths=0.4)
plt.xlabel('Cars model')
plt.ylabel('Cars price')
plt.title('Box plot for cars prices')
plt.xticks(ticks=[0,1,2],
           labels=['Software Eengineer','Software Eengineer Software','Software Eengineer Learning'],
           rotation=30
          )
plt.show()
```
