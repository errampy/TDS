```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.array([0,2,4,6,8])
y = np.array([1,20,10,50,30])
plt.plot(x,y)
plt.show()
```

```python
# visulize the sine 
angle = np.radians(np.arange(0,1441))
sine = np.sin(angle)
cosine = np.cos(angle)
```

```python
# visualize this
plt.plot(angle, sine, c='red')
plt.plot(angle, cosine, c='green')
plt.show()
```
