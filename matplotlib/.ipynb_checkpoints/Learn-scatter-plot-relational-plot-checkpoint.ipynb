# Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./pd/DataSets/usedcars.csv')
df.head()
```

```python
plt.scatter(x='price', y='mileage', data=df)
plt.show()
```

# How to add X Label

```python
plt.scatter(x='price', y='mileage', data=df)
plt.xlabel('Car prices')# add the x label
plt.show()
```

# How To Add Y Label

```python
plt.scatter(x='price', y='mileage', data=df)
plt.ylabel('Car milages')# add the y label
plt.show()
```

# How to add title

plt.scatter(x='price', y='mileage', data=df)
plt.title('Scatter Ploat') # add the title
plt.show()

# How to change Color

```python
plt.scatter(x='price', y='mileage', data=df,c='red')
plt.show()
```

# Combine All Together

```python
plt.scatter(x='price', y='mileage', data=df, c='blue')
plt.xlabel('Car Prices')
plt.ylabel('Car Milages')
plt.title('Scatter Plot of used cars dataset')
plt.show()
```

# How to add marker style
'o' == circle
's' == square
+
^

```python
plt.scatter(x='price', y='mileage',data=df, c='green',marker='D')
# marker = p, s, H, d, D, o....
plt.show()
```

# How filter the data

```python
plt.scatter(x='price', y='mileage', data=df[df.model=='SE'], c='red')
plt.scatter(x='price', y='mileage', data=df[df.model=='SEL'], c='green')
plt.scatter(x='price', y='mileage', data=df[df.model=='SES'],c='blue')
plt.xlabel('Car Prices')
plt.ylabel('Car Mileages')
plt.title('Scatter - Used Cars')
plt.show()
                                            
```

# How to add Legend and Label

```python
plt.scatter(x='price', y='mileage', data=df[df.model=='SE'], c='red', marker='d', label='SE')
plt.scatter(x='price', y='mileage', data=df[df.model=='SEL'], c='green', marker='p', label='SEL')
plt.scatter(x='price', y='mileage', data=df[df.model=='SES'],c='blue', marker='+', label='SES')
plt.xlabel('Car Prices')
plt.ylabel('Car Mileages')
plt.title('Scatter - Used Cars')
plt.legend()
plt.show()
```

# Customisation of Ledend

```python
plt.scatter(x='price', y='mileage', data=df[df.model=='SE'], c='red', marker='d', label='SE')
plt.scatter(x='price', y='mileage', data=df[df.model=='SEL'], c='green', marker='p', label='SEL')
plt.scatter(x='price', y='mileage', data=df[df.model=='SES'],c='blue', marker='+', label='SES')
plt.xlabel('Car Prices')
plt.ylabel('Car Mileages')
plt.title('Scatter - Used Cars')
plt.legend(loc='best', title='cars models', ncols=1, fancybox=False, shadow=True) # best, upper left, upper right, upper center
plt.show()
```
